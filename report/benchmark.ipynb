```python
import matplotlib.pyplot as plt
import numpy as np
from os import listdir
from os.path import isfile, join

mat = []
with open("data/taai11-300.txt") as f:
    lines = f.readlines()
    for line in lines:
        mat.append([float(x) for x in line.split(',')])

mat = np.array(mat)
print(max(mat[:,3])/625, min(mat[:,3])/625)

datafiles = [f for f in listdir("data/") if isfile(join("data/", f)) and f.endswith(".txt") and '300' in f]
datafiles.sort()
series = []
files = []
for file in datafiles:
    if file == "taai11-300.txt":
        continue
    files.append(file)
    print(file)
    with open("data/" + file) as f:
        lines = f.readlines()
        arr = np.array([float(line.strip()) for line in lines if float(line.strip()) <= 60.0], dtype=np.float64)
        series.append(arr)
        filtered_data = arr[arr < 60.0]
        print(file.split(".")[0].split("300-")[-1], "num:", len(filtered_data), "avg:", "total:", sum(arr), sum(filtered_data)/len(filtered_data), "median:", np.median(filtered_data))
        

sorted_series = [np.sort(arr) for arr in series]
sorted_series[0], sorted_series[7] = sorted_series[7], sorted_series[0]
sorted_series[1], sorted_series[6] = sorted_series[6], sorted_series[1]
sorted_series[7], sorted_series[6] = sorted_series[6], sorted_series[7]
sorted_series[0], sorted_series[2] = sorted_series[2], sorted_series[0]
sorted_series[1], sorted_series[8] = sorted_series[8], sorted_series[1]
files[0], files[7] = files[7], files[0]
files[1], files[6] = files[6], files[1]
files[7], files[6] = files[6], files[7]
files[0], files[2] = files[2], files[0]
files[1], files[8] = files[8], files[1]

fig, ax = plt.subplots()
fig.set_figwidth(9)
fig.set_figheight(7)
for i, (s, file) in enumerate(zip(sorted_series, files)):
    emph = ["benchmark_taai11-300-symbolic-block-start.txt", "benchmark_taai11-300-sbs-improved.txt", "benchmark_taai11-300-brute-force.txt"]
    width = 2.5 if file in emph else 1.5
    offset = 0 if not 'bgu' in file else 3
    colour = f"C{i+offset}" if not 'aot' in file else f"C{i+offset-1}"
    colour = 'purple' if 'pbnsolve' in file else colour
    style = f"-" if not 'aot' in file else f"--"
    plt.plot(s[s<60.0], label = file.split(".")[0].split("300-")[-1], linewidth=width, color=colour, linestyle=style)


ax.set_ylim(1e-1, 60.0)
ax.set_yscale("log")
# ax.set_xscale("log")
plt.xlabel('Nonograms decided')
plt.ylabel('Runtime $\\left[s\\right]$')
plt.title('Time to decide Uniqueness (TAAI11, difficult 25x25, $0.34 < \\rho < 0.46$), 60s timeout')
plt.legend(loc='lower right')
plt.grid(True)
plt.savefig("benchmark_taai11_300.pdf")
plt.close()
```
